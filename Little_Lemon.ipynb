```python
query = """
SELECT
    Customer.CustomerID,
    Customer.CustomerName,
    City.City,
    PostalCode.PostalCode,
    Country.Country,
    Orders.Cost
FROM Customer
LEFT JOIN Address
ON Customer.AddressID = Address.AddressID
LEFT JOIN City
ON Address.CityID = City.CityID
LEFT JOIN PostalCode
ON Address.PostalCodeID = PostalCode.PostalCodeID
LEFT JOIN Country
ON Address.CountryID = Country.CountryID
LEFT JOIN Orders
ON Orders.CustomerID = Orders.CustomerID
WHERE Cost > 60
"""

cursor.execute(query)
```

```python
results = cursor.fetchall()
print(results)
```
